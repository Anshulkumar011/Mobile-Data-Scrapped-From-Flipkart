```python
import requests
from bs4 import BeautifulSoup
import csv
import time
import random

# Flipkart base URL
search_url = "https://www.flipkart.com/search?q=mobile+under+50000"

# Fake browser headers (bot block na ho)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0 Safari/537.36"
}

# CSV file create
file = open("mobile.csv", "w", newline="", encoding="utf-8")
writer = csv.writer(file)
writer.writerow(["Product Name", "Price", "Rating", "Storage", "Camera", "Display", "Battery"])

# Loop pages
for page in range(1, 543):  # 5 pages scrape karenge
    print(f"Scraping page {page}...")
    url = f"{search_url}&page={page}"

    try:
        r = requests.get(url, headers=headers, timeout=15)
        r.raise_for_status()
        soup = BeautifulSoup(r.text, "lxml")

        # Products ka data nikalna
        names = soup.find_all("div", class_="KzDlHZ")
        prices = soup.find_all("div", class_="Nx9bqj")
        ratings = soup.find_all("div", class_="XQDdHH")

        # Features wali list (isme Storage, Camera, Battery, Display hote hain)
        features = soup.find_all("ul", class_="G4BRas")

        # Har product ke liye loop
        for n, p, r_, f in zip(names, prices, ratings, features):
            name = n.get_text(strip=True)
            price = p.get_text(strip=True)
            rating = r_.get_text(strip=True)

            # ul ke andar saare <li> hain (storage, camera, battery, display...)
            li_tags = f.find_all("li")

            # Default blank values
            storage = ""
            camera = ""
            display = ""
            battery = ""

            # Features check karo
            for li in li_tags:
                text = li.get_text(strip=True)
                if "RAM" in text or "GB" in text:   # storage
                    storage = text
                elif "MP" in text:  # camera
                    camera = text
                elif "Display" in text:  # display
                    display = text
                elif "mAh" in text:  # battery
                    battery = text

            # CSV me write kar do
            writer.writerow([name, price, rating, storage, camera, display, battery])

        # Random delay taaki Flipkart block na kare
        time.sleep(random.uniform(2, 5))

    except Exception as e:
        print("Error:", e)

file.close()
print("✅ Data saved in mobiles.csv")
```
